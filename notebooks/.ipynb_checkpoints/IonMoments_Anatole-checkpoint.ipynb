# Inversion of moments skew-normal


```python
import sys
# sys.path.append('../tools/')
# import Functions_Heavy as Fn
import sympy as sym
import numpy as np
import time
```

```python
def sumation_term_W_1_6(k, a, a1, beta):
    total_sum = 0
    for l in range(k+1):
        term = (a*beta)**(2*k - 2*l)/(4**l*sym.factorial(l)*sym.factorial(2*k-2*l)*(a**2 + a1)**(2*k - l))
        total_sum += term
    return total_sum

class W_1_6(sym.Function):
    
    @classmethod  
    def eval(cls, m, a, a1, beta):
        
        total_sum = 0
        for k in range(m+1):
            term = sym.factorial(2*k)/(sym.factorial(k)*a1**(m + 1 - k))*sumation_term_W_1_6(k, a, a1, beta)
            total_sum += term
            
        coefficient = sym.simplify(sym.exp(-a1*beta**2/(a**2 + a1)))

        return coefficient*total_sum
```

```python
def sumation_term_W_3_7(l, a, a1, beta):
    total_sum = 0
    for r in range(l+1):
        term = a1**(2*l + 2 - 2*r)*(2*beta)**(2*l + 1 - 2*r)/(sym.factorial(r)*sym.factorial(2*l + 1 - 2*r)*(a + a1**2)**(2*l + 1 - r))
        total_sum += term
    return total_sum

class W_3_7(sym.Function):
    
    @classmethod  
    def eval(cls, m, a, a1, beta):
        
        total_sum = 0
        for l in range(m):
            term = sym.factorial(l)/a**(m - l)*sumation_term_W_3_7(l, a, a1, beta)
            total_sum += term
            
        coefficient = 1/(sym.factorial(m)*sym.sqrt(a + a1**2))*sym.simplify(sym.exp(-a*beta**2/(a + a1**2)))
        
        first_term = sym.sqrt(sym.pi)/(sym.factorial(m)*a**m*sym.sqrt(a))*sym.erf(sym.sqrt(a)*beta/(sym.sqrt(a + a1**2)))
        
        coefficient = sym.simplify(coefficient)
        total_sum = sym.simplify(total_sum)
        first_term = sym.simplify(first_term)

        return first_term - coefficient*total_sum

```

```python
def find_cross_x(X, A, x):
    if x is None:
       return None
    it = np.arange(0, len(A), 1, dtype=int)
    if (X==x).any():
       return A[X==x][0]
    if x<X.min():
       it_m = 0
    elif x>X.max():
       it_m = -1
    else:
       it_m = it[X<x][-1]
    it_p = it_m + 1
    coef = (X[it_p]-x)/(X[it_p]-X[it_m])
    return A[it_m]*coef + A[it_p]*(1-coef)



def computeDensity(v, ivdf):
    int_x = np.trapz(ivdf, x=v)    
    return int_x

def computeMoment(v, ivdf, order):
    int_x = np.trapz(ivdf*v**order, x=v)/np.trapz(ivdf, x=v)   
    return int_x

def computeVelocity(v, ivdf):
    int_x = np.trapz(ivdf*v, x=v)/np.trapz(ivdf, x=v)
    
    return int_x#.to_value(u.m/u.s)

def computeTxx(v, ivdf):
    int_x = np.trapz(ivdf*v*v, x=v)/np.trapz(ivdf, x=v)
    
    return int_x#.to_value(u.m**2/u.s**2)

def computeQ_xxx(v, ivdf):
    int_x = np.trapz(ivdf*v*v*v, x=v)/np.trapz(ivdf, x=v)
    
    return int_x#.to_value(u.m**3/u.s**3)

def computeR_xxxx(v, ivdf):
    int_x = np.trapz(ivdf*v*v*v*v, x=v)/np.trapz(ivdf, x=v)

    return int_x#.to_value(u.m**4/u.s**4)

x_0 = (view.parameters['Lx']/1001*(1001+x_id_0)).to_value(u.m)
x_1 = (view.parameters['Lx']/1001*(1001+x_id_1)).to_value(u.m)
x_2 = (view.parameters['Lx']/1001*(1001+x_id_2)).to_value(u.m)
x_3 = (view.parameters['Lx']/1001*(1001+x_id_3)).to_value(u.m)

u_0 = computeVelocity(v, ivdf_0)
u_1 = computeVelocity(v, ivdf_1)
u_2 = computeVelocity(v, ivdf_2)
u_3 = computeVelocity(v, ivdf_3)

u_av = computeVelocity(v, vdf_avg)
x_av = np.linspace(0, 0.1, np.shape(vdf)[1])

u_ivdf = [u_0, u_1, u_2, u_3]
x_ivdf = [x_0, x_1, x_2, x_3]
X      = view.x.to_value(u.m)

# n_0 = computeDensity(v, ivdf_0)
# u_i_0 = computeVelocity(v, ivdf_0)
# q_xxx_0 = computeQ_xxx(v, ivdf_0)
# T_xx_0  = computeTxx(v, ivdf_0)
# r_xxxx_0 = computeR_xxxx(v, ivdf_0)
# n_0
x_ivdf
```

```python
def PearsonIV_1D(vel, a, k, nu, lamda, m):
                     
    return 1./(a*k)*np.exp(-nu*np.arctan((vel - lamda)/a))/(1 + ((vel - lamda)/a)**2)**m
```

## Comparison to Maximum entropy

```python
from scipy.optimize import curve_fit

def MaxEntropy_1D(vel, a, b, c, d, e):
                     
    return a*np.exp(b*vel + c*vel**2 + d*vel**3 - abs(e)*vel**4)

def MaxEntropy7M_1D(vel, a, b, c, d, e, f, g):

    return a*np.exp(b*vel + c*vel**2 + d*vel**3 + e*vel**4 + f*vel**5 - abs(g)*vel**6)

def fitIVDF(func, v, ivdf, initialGuessArray):

    popt, pcov = curve_fit(func, v, ivdf, p0=initialGuessArray, maxfev = 1200000)
    return popt

def computeMomentArray(v, ivdf, order):
    Moments = np.zeros(order+1)
    n_0   = computeDensity(v, ivdf)
    u_i_0 = computeMoment(v, ivdf, 1)
    Moments[0] = n_0
    Moments[1] = u_i_0
    
    for i in range(order-1):
        Moments[i+2] = computeMoment((v - u_i_0), ivdf, i+2)

    return Moments

def NormalizedMomentArray(Moments):
    NormalizedMom    = np.zeros(len(Moments))
    NormalizedMom[0] = 1
    v_th_0           = np.sqrt(Moments[2])
    
    for i in range(len(Moments)-1):
        NormalizedMom[i+1] = Moments[i+1]/v_th_0**(i+1)

    return NormalizedMom
        
```

```python
# Compute moments and normalized moments
Moments = computeMomentArray(v, ivdf_0, 4)
Moments_Norm = NormalizedMomentArray(Moments)

n_0    = Moments[0]
v_th_0 = np.sqrt(Moments[2])

print("Moments :", Moments, ", Normalized:",Moments_Norm)
```

```python
def find_cross_x(X, A, x):
    if x is None:
       return None
    it = np.arange(0, len(A), 1, dtype=int)
    if (X==x).any():
       return A[X==x][0]
    if x<X.min():
       it_m = 0
    elif x>X.max():
       it_m = -1
    else:
       it_m = it[X<x][-1]
    it_p = it_m + 1
    coef = (X[it_p]-x)/(X[it_p]-X[it_m])
    return A[it_m]*coef + A[it_p]*(1-coef)



def computeDensity(v, ivdf):
    int_x = np.trapz(ivdf, x=v)    
    return int_x

def computeMoment(v, ivdf, order):
    int_x = np.trapz(ivdf*v**order, x=v)/np.trapz(ivdf, x=v)   
    return int_x

def computeVelocity(v, ivdf):
    int_x = np.trapz(ivdf*v, x=v)/np.trapz(ivdf, x=v)
    
    return int_x#.to_value(u.m/u.s)

def computeTxx(v, ivdf):
    int_x = np.trapz(ivdf*v*v, x=v)/np.trapz(ivdf, x=v)
    
    return int_x#.to_value(u.m**2/u.s**2)

def computeQ_xxx(v, ivdf):
    int_x = np.trapz(ivdf*v*v*v, x=v)/np.trapz(ivdf, x=v)
    
    return int_x#.to_value(u.m**3/u.s**3)

def computeR_xxxx(v, ivdf):
    int_x = np.trapz(ivdf*v*v*v*v, x=v)/np.trapz(ivdf, x=v)

    return int_x#.to_value(u.m**4/u.s**4)

x_0 = (view.parameters['Lx']/1001*(1001+x_id_0)).to_value(u.m)
x_1 = (view.parameters['Lx']/1001*(1001+x_id_1)).to_value(u.m)
x_2 = (view.parameters['Lx']/1001*(1001+x_id_2)).to_value(u.m)
x_3 = (view.parameters['Lx']/1001*(1001+x_id_3)).to_value(u.m)

u_0 = computeVelocity(v, ivdf_0)
u_1 = computeVelocity(v, ivdf_1)
u_2 = computeVelocity(v, ivdf_2)
u_3 = computeVelocity(v, ivdf_3)

u_ivdf = [u_0, u_1, u_2, u_3]
x_ivdf = [x_0, x_1, x_2, x_3]
X      = view.x.to_value(u.m)

# n_0 = computeDensity(v, ivdf_0)
# u_i_0 = computeVelocity(v, ivdf_0)
# q_xxx_0 = computeQ_xxx(v, ivdf_0)
# T_xx_0  = computeTxx(v, ivdf_0)
# r_xxxx_0 = computeR_xxxx(v, ivdf_0)
# n_0
x_ivdf
```

## Comparison to Maximum entropy

```python
from scipy.optimize import curve_fit

def MaxEntropy_1D(vel, a, b, c, d, e):
                     
    return a*np.exp(b*vel + c*vel**2 + d*vel**3 - abs(e)*vel**4)

def MaxEntropy7M_1D(vel, a, b, c, d, e, f, g):

    return a*np.exp(b*vel + c*vel**2 + d*vel**3 + e*vel**4 + f*vel**5 - abs(g)*vel**6)

def fitIVDF(func, v, ivdf, initialGuessArray):

    popt, pcov = curve_fit(func, v, ivdf, p0=initialGuessArray, maxfev = 1200000)
    return popt

def computeMomentArray(v, ivdf, order):
    Moments = np.zeros(order+1)
    n_0   = computeDensity(v, ivdf)
    u_i_0 = computeMoment(v, ivdf, 1)
    Moments[0] = n_0
    Moments[1] = u_i_0
    
    for i in range(order-1):
        Moments[i+2] = computeMoment((v - u_i_0), ivdf, i+2)

    return Moments

def NormalizedMomentArray(Moments):
    NormalizedMom    = np.zeros(len(Moments))
    NormalizedMom[0] = 1
    v_th_0           = np.sqrt(Moments[2])
    
    for i in range(len(Moments)-1):
        NormalizedMom[i+1] = Moments[i+1]/v_th_0**(i+1)

    return NormalizedMom
        
```
